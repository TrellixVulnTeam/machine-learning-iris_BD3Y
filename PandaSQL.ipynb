```python
import pandas as pd
from pandasql import sqldf
```

```python
df = pd.read_csv(r"C:\Users\gagan\Downloads\Job Task\Dataset\ipl_2022_dataset.csv", index_col=0)

df.head()
```

```python
df = df.rename(columns={"COST IN ₹ (CR.)":"cost_inr"})
```

```python
df = df.drop(columns='Cost IN $ (000)')
```

```python
mysql = lambda q: sqldf(q, globals())

mysql("SELECT * FROM df LIMIT 5;")
```

```python
mysql("""SELECT Player, cost_inr 
      FROM df
      WHERE TYPE = 'BATTER'
      ORDER BY 2 DESC
      LIMIT 3""")
```

```python
mysql("""SELECT Player, cost_inr 
      FROM df
      WHERE TYPE = 'BOWLER'
      ORDER BY 2 DESC
      LIMIT 5""")
```

```python
mysql("""SELECT Player, cost_inr 
      FROM df
      WHERE TYPE = 'ALL-ROUNDER'
      AND cost_inr is not null
      ORDER BY 2 DESC
      """)
```

```python
mysql("""SELECT TYPE, AVG(cost_inr) average_cost
      FROM df
      GROUP BY 1
      ORDER BY 2 DESC
      """)
```

```python
mysql("""SELECT Player, Team, cost_inr
      FROM df
      WHERE "Base Price" = 'Retained'
      ORDER BY 3 DESC
      """)
```

```python
mysql("""
With top_player as (
    SELECT TEAM, Player, cost_inr, ROW_NUMBER() OVER(PARTITION BY Team ORDER by cost_inr DESC) as rank
      FROM df
      WHERE cost_inr is not null)
      SELECT *
      FROM top_player
      WHERE rank = 1
      ORDER BY 3 DESC
      """)
```

```python
mysql("""SELECT *
      FROM df
      WHERE "Base Price" = 'Retained'
      ORDER BY 3 DESC
      """)
```

```python
df['New Base Price'] = df['Base Price'].replace({'20 Lakh':'0.2',
                                                 '50 Lakh':'0.5',
                                                 '2 Cr':'2.0',
                                                 '1 Cr':'1.0',
                                                 '75 Lakh':'0.75',
                                                 '30 Lakh':'0.3',
                                                 '40 Lakh':'0.4',
                                                 'Draft Pick':'0',
                                                 'Retained':'0',
                                                 '1.5 Cr': '1.5'})
```

```python
mysql("""SELECT Player, "New Base Price", cost_inr, cost_inr - "New Base Price" as price_hike
      FROM df
      WHERE "New Base Price" != 0
      AND cost_inr is not null
      ORDER BY 4 DESC
      Limit 10
      """)
```
